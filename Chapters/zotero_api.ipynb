```python
# import the necessary libraries
import pyzotero
import requests

# create a Zotero client object
zot = pyzotero.Zotero('<your_library_id>', '<your_api_key>')

# get all items from your library
items = zot.everything(zot.top())

# loop through each item and get its DOI
for item in items:
    doi = item['data']['DOI']
    
    # use the DOI to make a request to the Crossref API
    url = 'https://api.crossref.org/works/' + doi
    response = requests.get(url)
    data = response.json()
    
    # update the Zotero item with the info obtained from the Crossref API
    zot.update_item(item['key'], data)
```
