```python
import re
import bibtexparser

# Parse the BibTeX files
def parse_bibtex_file(filename):
    with open(filename, 'r') as bibfile:
        bib_db = bibtexparser.load(bibfile)
    return bib_db.entries

# Create a mapping of old keys to new keys based on a matching condition
def create_key_mapping(old_entries, new_entries):
    mapping = {}
    for old_entry in old_entries:
        for new_entry in new_entries:
            if old_entry["title"] == new_entry["title"]:  # Adjust the matching condition if needed
                mapping[old_entry["ID"]] = new_entry["ID"]
                break
    return mapping

old_entries = parse_bibtex_file("references.bib")
new_entries = parse_bibtex_file("Zotero.bib")
key_mapping = create_key_mapping(old_entries, new_entries)

# Replace the old citation keys in main.tex with the new citation keys
with open("main.tex", "r") as main_file:
    content = main_file.read()

for old_key, new_key in key_mapping.items():
    content = re.sub(rf"\\cite\{{{old_key}\}}", rf"\\cite{{{new_key}}}", content)

# Save the updated content back to main.tex
with open("main.tex", "w") as main_file:
    main_file.write(content)
```

```python
# import the necessary libraries
import pyzotero
import requests

# create a Zotero client object
zot = pyzotero.Zotero('<your_library_id>', '<your_api_key>')

# get all items from your library
items = zot.everything(zot.top())

# loop through each item and get its DOI
for item in items:
    doi = item['data']['DOI']
    
    # use the DOI to make a request to the Crossref API
    url = 'https://api.crossref.org/works/' + doi
    response = requests.get(url)
    data = response.json()
    
    # update the Zotero item with the info obtained from the Crossref API
    zot.update_item(item['key'], data)
```
